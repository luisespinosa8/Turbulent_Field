# TGK formalism for computation of diffusion coefficients

```python
from pylab import *
import scipy.integrate
from scipy.optimize import minimize
from scipy import interpolate
from scipy.optimize import curve_fit
from scipy.stats import linregress
import iminuit 
from iminuit import Minuit
from scipy.signal import find_peaks
```

For PeV particles, the time step of measurements was $t = 0.1 \, r_L / c \sim 2.22\times 10^9$

## Quantities

```python
logE = np.round(np.arange(-1.0,1.2,0.2),decimals=1)
#logE = np.round(np.arange(-1.0,2.2,0.2),decimals=1)
E0 = 10**logE * 1e15# Energy (eV)

# L = 8 Lmax
l_c = 2.728 * 3.086e+16 # in meters
Lmax = 12.8 * 3.086e+16 # in meters
Lmin = 0.2 * 3.086e+16 # in meters

# L = 8 lc
#l_c = 13.079 * 3.086e+16 # in meters
#Lmax = 64 * 3.086e+16 # in meters
```

```python
eta = 0.5 # Turbulence level
Z = 1 # Proton atomic number
c = 3e8 # Speed of light (m/s)
B_0 = 1e-6
dB = eta * B_0
B = np.sqrt(B_0**2 + dB**2) * 1e-4 # in T
```

```python
# Larmor radius
def r_L(E, B):
    # E in eV
    Z = 1
    c = 3e8
    r = E / (Z * B * c) # in m
    return r
```

## Correlation functions

$R_{ij}(t) = \left< v_i(0) v_j(t)\right>$ 

$R_{zz}(t) = \frac{c^2}{3} \, e^{-t/\tau_\parallel}$ 


$R_{xx}(t) = R_{yy}(t) = \frac{c^2}{3}\cos{\omega t} \, e^{-t/\tau_\perp}$


$R_{xy}(t) = -\frac{c^2}{3}\sin{\omega t} \, e^{-t/\tau_A}$

```python
def R_parallel(t, tau):
    return (c**2 / 3) * np.exp(- t / tau)

def R_perpendicular(t, energy, tau, omega):
    return (c**2 / 3) * np.cos(omega * t) * np.exp(- t / tau)

def R_antisymmetric(t, energy, tau, omega):
    return - (c**2 / 3) * np.sin(omega * t) * np.exp(- t / tau)
```

## Diffusion coefficients

```python
def D_parallel(tau):
    return (c**2 / 3) * tau

def D_perpendicular(energy, tau, omega):
    #omega = c / r_L(energy, B)
    return (c**2 / 3) * tau * (1 / (1 + (omega * tau)**2))

def D_antisymmetric(energy, tau, omega):
    #omega = c / r_L(energy, B)
    return (c**2 / 3) * omega * tau**2 * (1 / (1 + (omega * tau)**2))
```

## Fitting of correlation functions using minimize

```python
omega_perpendicular = np.zeros(len(logE))
omega_antisymmetric = np.zeros(len(logE))
tau_parallel = np.zeros(len(logE))
tau_perpendicular = np.zeros(len(logE))
tau_antisymmetric = np.zeros(len(logE))
tau_guess_parallel = 1e9
tau_guess_perpendicular = 5e8
tau_guess_antisymmetric = 5e8
fit_limits = np.round( np.arange(10., 2000., (2000. - 100.) / len(logE)) )
fit_limits = fit_limits.astype(int)
for i in range(0, len(logE)):
    energy = E0[i]
    time = np.genfromtxt('{Path}/Correlation_Functions/time_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Rxx = np.genfromtxt('{Path}/Correlation_Functions/Rxx_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Ryy = np.genfromtxt('{Path}/Correlation_Functions/Ryy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Rzz = np.genfromtxt('{Path}/Correlation_Functions/Rzz_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Rxy = np.genfromtxt('{Path}/Correlation_Functions/Rxy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Varxx = np.genfromtxt('{Path}/Variance/Varxx_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta)) 
    Varyy = np.genfromtxt('{Path}/Variance/Varyy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta)) 
    Varzz = np.genfromtxt('{Path}/Variance/Varzz_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta)) 
    Varxy = np.genfromtxt('{Path}/Variance/Varxy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta)) 
    sigmaxx = np.sqrt(Varxx) / sqrt(5e3)
    sigmayy = np.sqrt(Varyy) / sqrt(5e3)
    sigmazz = np.sqrt(Varzz) / sqrt(5e3)
    sigmaxy = np.sqrt(Varxy) / sqrt(5e3)
    
    omega_guess = c / r_L(energy, B)
    
    index_zero = np.where(np.logical_and(Rzz < 0, Rzz < 0))
    fit_limit = min(index_zero[0])
    fit_limit2 = fit_limits[i]

    def R_parallel_fit(tau):
        function = R_parallel(time[0:fit_limit], tau)
        simulation = Rzz[0:fit_limit]
        Sigma = sigmazz[0:fit_limit]
        return sum( (function - simulation)**2 / Sigma**2 )
    output_parallel = minimize(R_parallel_fit, tau_guess_parallel, method='Nelder-Mead')
    print('Success (parallel, 10tothe{Elog} PeV):'.format(Elog = logE[i]), output_parallel['success'])
    fit_parallel = output_parallel['x']
    tau_parallel[i] = fit_parallel[0]
    
    def R_perpendicular_fit(parameters):
        tau, omega = parameters
        function = R_perpendicular(time[0:fit_limit2], energy, tau, omega) * (3 / c**2)
        simulation = Rxx[0:fit_limit2] * (3 / c**2)
        Sigma = sigmaxx[0:fit_limit2]
        return sum( (function - simulation)**2 / Sigma**2 )
    output_perpendicular = minimize(R_perpendicular_fit, (tau_guess_perpendicular, omega_guess), method='Nelder-Mead')
    print('Success (perpendicular, 10tothe{Elog} PeV):'.format(Elog = logE[i]), output_perpendicular['success'])
    fit_perpendicular = output_perpendicular['x']
    tau_perpendicular[i] = fit_perpendicular[0]
    omega_perpendicular[i] = fit_perpendicular[1]
    
    def R_antisymmetric_fit(parameters):
        tau, omega = parameters
        function = R_antisymmetric(time[0:fit_limit2], energy, tau, omega) * (3 / c**2)
        simulation = Rxy[0:fit_limit2] * (3 / c**2)
        Sigma = sigmaxy[0:fit_limit2]
        return sum( (function - simulation)**2 / Sigma**2 )
    output_antisymmetric = minimize(R_antisymmetric_fit, (tau_guess_antisymmetric, omega_guess), method='Nelder-Mead')
    print('Success (antisymmetric, 10tothe{Elog} PeV):'.format(Elog = logE[i]), output_antisymmetric['success'])
    fit_antisymmetric = output_antisymmetric['x']
    tau_antisymmetric[i] = fit_antisymmetric[0]
    omega_antisymmetric[i] = fit_antisymmetric[1]
    
    tau_guess_parallel = tau_parallel[i]
    tau_guess_perpendicular = tau_perpendicular[i]
    tau_guess_antisymmetric = tau_antisymmetric[i] 
    
tau_parallel = tau_parallel[2::]
tau_perpendicular = tau_perpendicular[2::]
omega_perpendicular = omega_perpendicular[2::]
tau_antisymmetric = tau_antisymmetric[2::]
omega_antisymmetric = omega_antisymmetric[2::]
```

## Fitting of correlation functions using minuit

```python
def find_first_peaks(x, y, npeaks):
    """
    Identify the first N peaks in the data.
    
    Parameters:
        x (np.ndarray): Array of x values.
        y (np.ndarray): Array of y values.
        npeaks (int): Number of peaks to identify (default: 3).
    
    Returns:
        tuple: (peak_positions, first_peaks)
    """
    peaks, _ = find_peaks(y)
    
    if len(peaks) < npeaks:
        print(f"Warning: Only {len(peaks)} peaks found, expected {npeaks}.")
        npeaks = len(peaks)
    
    first_peaks = peaks[:npeaks]
    peak_positions = x[first_peaks]
    peaks = y[first_peaks]
    
    return peak_positions, peaks, first_peaks
```

```python
# Fit parameters using Minuit
tau_parallel = np.zeros(len(logE))
tau_perpendicular = np.zeros(len(logE))
tau_antisymmetric = np.zeros(len(logE))
omega_perpendicular = np.zeros(len(logE))
omega_antisymmetric = np.zeros(len(logE))
# Errors
error_parallel = np.zeros(len(logE))
error_perpendicular = np.zeros(len(logE))
error_antisymmetric = np.zeros(len(logE))
error_omega_perpendicular = np.zeros(len(logE))
error_omega_antisymmetric = np.zeros(len(logE))
# Initial guesses
tau_guess_parallel = 1e9
tau_guess_perpendicular = 5e8
tau_guess_antisymmetric = 5e8
fit_limits = np.round( np.arange(10., 2000., (2000. - 100.) / len(logE)) )
fit_limits = fit_limits.astype(int)
for i in range(0, len(logE)):
    energy = E0[i]
    time = np.genfromtxt('{Path}/Correlation_Functions/time_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Rxx = np.genfromtxt('{Path}/Correlation_Functions/Rxx_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Ryy = np.genfromtxt('{Path}/Correlation_Functions/Ryy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Rzz = np.genfromtxt('{Path}/Correlation_Functions/Rzz_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Rxy = np.genfromtxt('{Path}/Correlation_Functions/Rxy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta))
    Varxx = np.genfromtxt('{Path}/Variance/Varxx_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta)) 
    Varyy = np.genfromtxt('{Path}/Variance/Varyy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta)) 
    Varzz = np.genfromtxt('{Path}/Variance/Varzz_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta)) 
    Varxy = np.genfromtxt('{Path}/Variance/Varxy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Path = path, Elog = logE[i], turb = eta)) 
    sigmaxx = np.sqrt(Varxx) / sqrt(5e3)
    sigmayy = np.sqrt(Varyy) / sqrt(5e3)
    sigmazz = np.sqrt(Varzz) / sqrt(5e3)
    sigmaxy = np.sqrt(Varxy) / sqrt(5e3)
    
    omega_guess = c / r_L(energy, B)
    peak_positions, peaks, _ = find_first_peaks(x, y, 10)

    distance_peaks = peak_positions[1] - peak_positions[0]
    omega = 2 * math.pi / distance_peaks
    fit = np.polyfit(peak_positions, np.log(peaks), 1)
    tau = -fit[0]
    
    index_zero = np.where(np.logical_and(Rzz < 0, Rzz < 0))
    fit_limit = min(index_zero[0])
    fit_limit2 = fit_limits[i]

    def R_parallel_fit(tau):
        function = R_parallel(time[0:fit_limit], tau)
        simulation = Rzz[0:fit_limit]
        Sigma = sigmazz[0:fit_limit]
        return sum( (function - simulation)**2 / Sigma**2 )
    output_parallel = Minuit(R_parallel_fit, tau_guess_parallel)
    output_parallel.errordef = Minuit.LEAST_SQUARES
    output_parallel.simplex()
    output_parallel.migrad()
    output_parallel.hesse()
    tau_parallel[i] = output_parallel.values[0]
    error_parallel[i] = output_parallel.errors[0]
    
    def R_perpendicular_fit(tau, omega):
        function = R_perpendicular(time[0:fit_limit2], energy, tau, omega) * (3 / c**2)
        simulation = Rxx[0:fit_limit2] * (3 / c**2)
        Sigma = sigmaxx[0:fit_limit2]
        return sum( (function - simulation)**2 / Sigma**2 )
    output_perpendicular = Minuit(R_perpendicular_fit, tau_guess_parallel, omega_guess)
    #output_perpendicular.errordef = Minuit.LEAST_SQUARES
    output_perpendicular.simplex()
    output_perpendicular.migrad()
    output_perpendicular.hesse()
    tau_perpendicular[i] = output_perpendicular.values[0]
    omega_perpendicular[i] = output_perpendicular.values[1]
    error_perpendicular[i] = output_perpendicular.errors[0]
    error_omega_perpendicular[i] = output_perpendicular.errors[1]
    
    def R_antisymmetric_fit(tau, omega):
        function = R_antisymmetric(time[0:fit_limit2], energy, tau, omega) * (3 / c**2)
        simulation = Rxy[0:fit_limit2] * (3 / c**2)
        Sigma = sigmaxy[0:fit_limit2]
        return sum( (function - simulation)**2 / Sigma**2 )
    output_antisymmetric = Minuit(R_antisymmetric_fit, tau_guess_antisymmetric, omega_guess)
    output_antisymmetric.errordef = Minuit.LEAST_SQUARES
    output_antisymmetric.simplex()
    output_antisymmetric.migrad()
    output_antisymmetric.hesse()
    tau_antisymmetric[i] = output_antisymmetric.values[0]
    omega_antisymmetric[i] = output_antisymmetric.values[1]
    error_antisymmetric[i] = output_antisymmetric.errors[0]
    error_omega_antisymmetric[i] = output_antisymmetric.errors[1]
    
    tau_guess_parallel = tau_parallel[i]
    tau_guess_perpendicular = tau_perpendicular[i]
    tau_guess_antisymmetric = tau_antisymmetric[i] 
    
    
tau_parallel = tau_parallel[2::]
tau_perpendicular = tau_perpendicular[2::]
omega_perpendicular = omega_perpendicular[2::]
tau_antisymmetric = tau_antisymmetric[2::]
omega_antisymmetric = omega_antisymmetric[2::]
error_parallel = error_parallel[2::]
error_perpendicular = error_perpendicular[2::]
error_antisymmetric = error_antisymmetric[2::]
```

```python
print('tau_para:', tau_parallel)
print('tau_perp:', tau_perpendicular)
print('tau_anti:', tau_antisymmetric)
```

## Computation diffusion coefficients

```python
Dzz = D_parallel(tau_parallel)
Dxx = D_perpendicular(E0, tau_perpendicular, omega_perpendicular)
Dxy = D_antisymmetric(E0, tau_antisymmetric, omega_antisymmetric)

ratio = r_L(E0[2::], B) / l_c
```

## Plotting

### Parallel coefficient

```python
# Plot of parallel diffusion coefficient and comparison with mean displacement method
Dzz_displacement = np.genfromtxt('Diffusion_Coefficients/Displacement_Method/Dzz_eta0.5.txt')
VarDzz_displacement = np.genfromtxt('Diffusion_Coefficients/Displacement_Method/VarDzz_eta0.5.txt') 
sigmazz = np.sqrt(VarDzz_displacement[2::]) / sqrt(1e3)

error_parallel_TGK = (error_parallel / tau_parallel) * Dzz
Dzz_displacement = Dzz_displacement[7:16]

alpha = - 1./3.
plt.figure(figsize=(10, 7))
plt.plot(ratio, Dzz / (l_c * c) * ratio**alpha, '-o', color='blue', label=r'$D_{zz}$ by TGK method')
plt.fill_between(ratio, (Dzz - error_parallel_TGK) / (l_c * c) * ratio**alpha, (Dzz + error_parallel_TGK) / (l_c * c) * ratio**alpha, color='blue', alpha=0.2)
plt.plot(ratio, Dzz_displacement / (l_c * c) * ratio**alpha, '-o', color='red', label=r'$D_{zz}$ by displacement method')
plt.fill_between(ratio, (Dzz_displacement - sigmazz) / (l_c * c) * ratio**alpha, (Dzz_displacement + sigmazz) / (l_c * c) * ratio**alpha, color='red', alpha=0.2)
plt.axvline(x = Lmin / l_c, linestyle='--', color='black')
plt.axvline(x = Lmax / l_c, linestyle='--', color='black')
plt.loglog()
#plt.ylim(1e0, 1e2)
plt.xlabel(r'$r_L / \lambda_c$', fontsize=17)
plt.ylabel(r'$D_{ij}$ / $c \lambda_c \times r_L^{-1/3}$', fontsize=17)
plt.legend(ncol=1, fontsize=17)
plt.title(r'Parallel diffusion coefficients of protons in turbulent field with $\eta =$ {turb}'.format(turb = eta), fontsize=17)
plt.savefig('Plots/TGK/comparison_displacement_parallel.pdf')
```
