# TGK formalism for computation of diffusion coefficients

```python
from pylab import *
import scipy.integrate
from scipy.optimize import minimize
from scipy import interpolate
from scipy.optimize import curve_fit
from scipy.stats import linregress
```

For PeV particles, the time step of measurements was $t = 0.1 \, r_L / c \sim 2.22\times 10^9$

## Quantities

```python
logE = np.round(np.arange(-1.0,1.6,0.2),decimals=1)
#logE = np.round(np.arange(-1.0,2.2,0.2),decimals=1)
E0 = 10**logE * 1e15# Energy (eV)

l_c = 2.728 * 3.086e+16 # in meters
#l_c = 13.079 * 3.086e+16 # in meters

Lmax = 12.8 * 3.086e+16 # in meters
#Lmax = 64 * 3.086e+16 # in meters
```

```python
eta = 0.5 # Turbulence level
Z = 1 # Proton atomic number
c = 3e8 # Speed of light (m/s)
B_0 = 1e-6
dB = eta * B_0
B = np.sqrt(B_0**2 + dB**2) * 1e-4 # in T
Lmin = 0.2 * 3.086e+16 # in meters
```

```python
# Larmor radius
def r_L(E, B):
    # E in eV
    Z = 1
    c = 3e8
    r = E / (Z * B * c) # in m
    return r
```

## Correlation functions

$R_{ij}(t) = \left< v_i(0) v_j(t)\right>$ 

$R_{zz}(t) = \frac{c^2}{3} \, e^{-t/\tau_\parallel}$ 


$R_{xx}(t) = R_{yy}(t) = \frac{c^2}{3}\cos{\omega t} \, e^{-t/\tau_\perp}$


$R_{xy}(t) = -\frac{c^2}{3}\sin{\omega t} \, e^{-t/\tau_A}$

```python
def R_parallel(t, tau):
    return (c**2 / 3) * np.exp(- t / tau)

def R_perpendicular(t, energy, tau, omega):
    return (c**2 / 3) * np.cos(omega * t) * np.exp(- t / tau)

def R_antisymmetric(t, energy, tau, omega):
    return - (c**2 / 3) * np.sin(omega * t) * np.exp(- t / tau)
```

## Diffusion coefficients

```python
def D_parallel(tau):
    return (c**2 / 3) * tau

def D_perpendicular(energy, tau, omega):
    #omega = c / r_L(energy, B)
    return (c**2 / 3) * tau * (1 / (1 + (omega * tau)**2))

def D_antisymmetric(energy, tau, omega):
    #omega = c / r_L(energy, B)
    return (c**2 / 3) * omega * tau**2 * (1 / (1 + (omega * tau)**2))
```

## Fitting of correlation functions

```python
omega_perpendicular = np.zeros(len(logE))
omega_antisymmetric = np.zeros(len(logE))
tau_perpendicular = np.zeros(len(logE))
tau_antisymmetric = np.zeros(len(logE))
tau_guess_perpendicular = 5e8
tau_guess_antisymmetric = 5e8
for i in range(0, len(logE)):
    energy = E0[i]
    time = np.genfromtxt('Data_5e3particles/L=8Lmax/Correlation_Functions/time_10tothe{Elog}_PeV_eta{turb}.txt'.format(Elog = logE[i], turb = eta))
    Rxx = np.genfromtxt('Data_5e3particles/L=8Lmax/Correlation_Functions/Rxx_10tothe{Elog}_PeV_eta{turb}.txt'.format(Elog = logE[i], turb = eta))
    Ryy = np.genfromtxt('Data_5e3particles/L=8Lmax/Correlation_Functions/Ryy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Elog = logE[i], turb = eta))
    Rxy = np.genfromtxt('Data_5e3particles/L=8Lmax/Correlation_Functions/Rxy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Elog = logE[i], turb = eta))
    Varxx = np.genfromtxt('Data_5e3particles/L=8Lmax/Variance/Varxx_10tothe{Elog}_PeV_eta{turb}.txt'.format(Elog = logE[i], turb = eta)) / sqrt(5e3)
    Varyy = np.genfromtxt('Data_5e3particles/L=8Lmax/Variance/Varyy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Elog = logE[i], turb = eta)) / sqrt(5e3)
    Varxy = np.genfromtxt('Data_5e3particles/L=8Lmax/Variance/Varxy_10tothe{Elog}_PeV_eta{turb}.txt'.format(Elog = logE[i], turb = eta)) / sqrt(5e3)
    sigmaxx = np.sqrt(Varxx) / sqrt(5e3)
    sigmayy = np.sqrt(Varyy) / sqrt(5e3)
    sigmaxy = np.sqrt(Varxy) / sqrt(5e3)
    
    omega_guess = c / r_L(energy, B)
    
    index_zero = np.where(np.logical_and(Rzz < 0, Rzz < 0))
    fit_limit = min(index_zero[0])
    fit_limit2 = 2000
    
    def R_perpendicular_fit(parameters):
        tau, omega = parameters
        function = R_perpendicular(time[0:fit_limit2], energy, tau, omega) * (3 / c**2)
        simulation = Rxx[0:fit_limit2] * (3 / c**2)
        Sigma = sigmaxx[0:fit_limit2]
        return sum( (function - simulation)**2 / Sigma**2 )
    output_perpendicular = minimize(R_perpendicular_fit, (tau_guess_perpendicular, omega_guess), method='Nelder-Mead')
    print('Success (perpendicular, 10tothe{Elog} PeV):'.format(Elog = logE[i]), output_perpendicular['success'])
    fit_perpendicular = output_perpendicular['x']
    tau_perpendicular[i] = fit_perpendicular[0]
    omega_perpendicular[i] = fit_perpendicular[1]
    
    def R_antisymmetric_fit(parameters):
        tau, omega = parameters
        function = R_antisymmetric(time[0:fit_limit2], energy, tau, omega) * (3 / c**2)
        simulation = Rxy[0:fit_limit2] * (3 / c**2)
        Sigma = sigmaxy[0:fit_limit2]
        return sum( (function - simulation)**2 / Sigma**2 )
    output_antisymmetric = minimize(R_antisymmetric_fit, (tau_guess_antisymmetric, omega_guess), method='Nelder-Mead')
    print('Success (antisymmetric, 10tothe{Elog} PeV):'.format(Elog = logE[i]), output_antisymmetric['success'])
    fit_antisymmetric = output_antisymmetric['x']
    tau_antisymmetric[i] = fit_antisymmetric[0]
    omega_antisymmetric[i] = fit_antisymmetric[1]
    
    tau_guess_parallel = tau_parallel[i]
    tau_guess_perpendicular = tau_perpendicular[i]
    tau_guess_antisymmetric = tau_antisymmetric[i] 
```

```python
print('tau_perp:', tau_perpendicular)
print('tau_anti:', tau_antisymmetric)
```

## Computation diffusion coefficients

```python
Dzz = D_parallel(tau_parallel)

Dxx = D_perpendicular(E0, tau_perpendicular, omega_perpendicular)
Dxy = D_antisymmetric(E0, tau_antisymmetric, omega_antisymmetric)
ratio = r_L(E0, B) / l_c

# Small test: according to Candia, for low turbulence level, all decorrelation timescales are the same
#Dxx = D_perpendicular(E0, tau_parallel)
#Dxy = D_antisymmetric(E0, tau_parallel)
```
