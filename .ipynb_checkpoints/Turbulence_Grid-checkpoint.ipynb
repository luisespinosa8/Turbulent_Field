# Grid generation following "SimpleGridTurbulence" module from CRPropa

```python
from pylab import *
import random
from scipy.stats import linregress
```

```python
n = 256
index = 5./3.
spacing = 100
Brms = 5
lmax = (n * spacing) / 8
lmin = 2 * spacing

alpha = -index -2
```

```python
n2 = int(floor(n / 2) + 1)

Bkx = np.zeros((n * n * n2),dtype=np.complex_)
Bky = np.zeros((n * n * n2),dtype=np.complex_)
Bkz = np.zeros((n * n * n2),dtype=np.complex_)
k0 = np.zeros((n * n * n2))

random.seed(42)

# N_grids possible discrete wave numbers
K = [0] * n
for i in range(0,n):
    K[i] = i / n - i / (n / 2.)

# Min and max k values
k_min = spacing / lmax
k_max = spacing / lmin

n0 = np.array((1,1,1))

# Computing Bkx, Bky and Bkz at each point of the grid
for ix in range(0,n):
    for iy in range(0,n):
        for iz in range(0,n2):
            i = ix * n * n2 + iy * n2 + iz
            ek = np.array((K[ix], K[iy], K[iz]))
            k = np.sqrt(K[ix]**2 + K[iy]**2 + K[iz]**2)
            k0[i] = k
            if k > k_max or k < k_min:
                Bkx[i] = 0;
                Bky[i] = 0;
                Bkz[i] = 0;
                continue
            if K[ix] == K[iy] == K[iz]:
                e1 = np.array((-1., 1., 0))
                e2 = np.array((1., 1., -2.))
            else:
                e1 = np.cross(ek, n0)
                e2 = np.cross(ek, e1)
            e1 = e1 / np.sqrt(e1[0]**2 + e1[1]**2 + e1[2]**2)
            e2 = e2 / np.sqrt(e2[0]**2 + e2[1]**2 + e2[2]**2)
            theta = 2 * np.pi * np.random.rand()
            b = e1 * np.cos(theta) + e2 * np.sin(theta)
            b = b * np.random.normal() * pow(k, alpha / 2)
            phase = 2 * np.pi * np.random.rand()
            cosPhase = np.cos(phase)
            sinPhase = np.sin(phase)
            Bkx[i] = (b[0] * cosPhase) + (b[0] * sinPhase)*1j
            Bky[i] = (b[1] * cosPhase) + (b[1] * sinPhase)*1j
            Bkz[i] = (b[2] * cosPhase) + (b[2] * sinPhase)*1j
```

```python
print('k_min:', k_min)
print('k_max:', k_max)
```

```python
# Inverse FFT to go to x-space
Bk = np.array((Bkx, Bky, Bkz))
B = np.fft.ifftn(Bk)
Bx = B[0]
By = B[1]
Bz = B[2]
```

```python
Bkx_mag = np.sqrt(Bkx.real**2 + Bkx.imag**2)
Bky_mag = np.sqrt(Bky.real**2 + Bky.imag**2)
Bkz_mag = np.sqrt(Bkz.real**2 + Bkz.imag**2)
#Bkx_mag = Bkx.real
#Bky_mag = Bky.real
#Bkz_mag = Bkz.real
Bk_mag = np.sqrt(Bkx_mag**2 + Bky_mag**2 + Bkz_mag**2)
print('Min Bk:', min(Bk_mag))
print('Max Bk:', max(Bk_mag))
```

## Plotting B(k) components

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 7))
counts, bins = np.histogram(Bkx_mag, bins=50)
ax[0,0].hist(bins[:-1], bins, weights=counts)
ax[0,0].set_xlabel(r'$\delta B_{k,x}$ [nG]', fontsize=12)
ax[0,0].grid()
counts, bins = np.histogram(Bky_mag, bins=50)
ax[0,1].hist(bins[:-1], bins, weights=counts)
ax[0,1].set_xlabel(r'$\delta B_{k,y}$ [nG]', fontsize=12)
ax[0,1].grid()
counts, bins = np.histogram(Bkz_mag, bins=50)
ax[1,0].hist(bins[:-1], bins, weights=counts)
ax[1,0].set_xlabel(r'$\delta B_{k,z}$ [nG]', fontsize=12)
ax[1,0].grid()
counts, bins = np.histogram(Bk_mag, bins=50)
ax[1,1].hist(bins[:-1], bins, weights=counts)
ax[1,1].set_xlabel(r'$\delta B_{k}$ [nG]', fontsize=12)
ax[1,1].grid()
plt.savefig('histogram_Bk.png')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 7))
counts, bins = np.histogram(Bkx_mag**2, bins=50)
ax[0,0].hist(bins[:-1], bins, weights=counts)
ax[0,0].set_xlabel(r'$\delta B_{k,x}^2$ [nG]', fontsize=8)
ax[0,0].grid()
ax[0,0].set_yscale('log')
counts, bins = np.histogram(Bky_mag**2, bins=50)
ax[0,1].hist(bins[:-1], bins, weights=counts)
ax[0,1].set_xlabel(r'$\delta B_{k,y}^2$ [nG]', fontsize=8)
ax[0,1].grid()
ax[0,1].set_yscale('log')
counts, bins = np.histogram(Bkz_mag**2, bins=50)
ax[1,0].hist(bins[:-1], bins, weights=counts)
ax[1,0].set_xlabel(r'$\delta B_{k,z}^2$ [nG]', fontsize=8)
ax[1,0].grid()
ax[1,0].set_yscale('log')
counts, bins = np.histogram(Bk_mag**2, bins=50)
ax[1,1].hist(bins[:-1], bins, weights=counts)
ax[1,1].set_xlabel(r'$\delta B_{k}^2$ [nG]', fontsize=8)
ax[1,1].grid()
ax[1,1].set_yscale('log')
plt.savefig('histogram_Bk^2.png')
```

```python
plt.figure(figsize=(10, 7))
plt.plot(k0, Bkx_mag**2, 'o', color='blue', label=r'$B_x$')
plt.plot(k0, Bky_mag**2, 'o', color='red', label=r'$B_y$')
plt.plot(k0, Bkz_mag**2, 'o', color='green', label=r'$B_z$')
plt.yscale('log')
plt.xlim(k_min, k_max)
plt.xlabel('k', fontsize=17)
plt.ylabel(r'$\delta B^2(k)$', fontsize=17)
plt.grid()
plt.legend(fontsize=14)
plt.savefig('B_k_components.png')
```

```python
counts, bins = np.histogram(np.log10(k0[1::]), bins=50)
logk_values = [0] * (len(bins) - 1) 
Bkx_sqrd_values = [0] * (len(bins) - 1) 
Bky_sqrd_values = [0] * (len(bins) - 1)
Bkz_sqrd_values = [0] * (len(bins) - 1)
Bk_sqrd_values = [0] * (len(bins) - 1)
for i in range(0, len(bins) - 1):
    logk_values[i] = (bins[i] + bins[i+1]) / 2
    index = np.where(np.logical_and(np.log10(k0) >= bins[i], np.log10(k0) <= bins[i+1]))
    Bkx_sqrd_values[i] = np.mean(Bkx_mag[index]**2)
    Bky_sqrd_values[i] = np.mean(Bky_mag[index]**2)
    Bkz_sqrd_values[i] = np.mean(Bkz_mag[index]**2)
    Bk_sqrd_values[i] = np.mean(Bk_mag[index]**2)

logk_values = np.array(logk_values)
Bkx_sqrd_values = np.array(Bkx_sqrd_values)
Bky_sqrd_values = np.array(Bky_sqrd_values)
Bkz_sqrd_values = np.array(Bkz_sqrd_values)
Bk_sqrd_values = np.array(Bk_sqrd_values)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(10**logk_values, Bkx_sqrd_values, color='blue', label=r'$\delta B_x$')
plt.plot(10**logk_values, Bky_sqrd_values, color='red', label=r'$\delta B_y$')
plt.plot(10**logk_values, Bkz_sqrd_values, color='green', label=r'$\delta B_z$')
plt.loglog()
#plt.xlim(k_min, k_max)
plt.xlabel('k', fontsize=17)
plt.ylabel(r'$\delta B^2(k)$', fontsize=17)
plt.legend(fontsize=14)
plt.savefig('Plots/spectrum_grid_components.png')
```

```python
plt.figure(figsize=(10, 7))
plt.plot(k0, Bk_mag**2, 'o', color='black')
plt.yscale('log')
plt.xlim(k_min, k_max)
plt.xlabel('k', fontsize=17)
plt.ylabel(r'$\delta B^2(k)$', fontsize=17)
plt.grid()
plt.savefig('B_k.png')
```

```python
plt.figure(figsize=(10, 7))
plt.plot(10**logk_values, Bk_sqrd_values, color='black')
plt.loglog()
plt.xlabel('k', fontsize=17)
plt.ylabel(r'$\delta B^2(k)$', fontsize=17)
plt.savefig('Plots/spectrum_grid.png')

# CHECK SLOPE
index_slope = np.where(np.logical_and(10**logk_values >= k_min, 10**logk_values <= k_max))
linear = linregress(logk_values[index_slope], np.log10(Bk_sqrd_values[index_slope]))
print('Slope of turbulence spectrum:', linear[0])
```
